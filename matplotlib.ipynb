```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
# Scatter plot
plt.scatter(X_data, y_data, c="#0000FF")
```

```python
# Line plot
years = [2006 + x for x in range(5)]

weights = [80, 85, 70, 92, 90]
plt.plot(years, weights, c="r", lw=2, ls="--")
```

```python
# Bar plot
plt.bar(years, weights, color='r', width=0.5,
        align='center', edgecolor='y', linewidth=2)


plt.title("bar charts")


plt.xlabel("year")


plt.ylabel("weights")
```

```python
# Histograms
ages = np.random.normal(20, 2.5, 1000)
plt.hist(ages, bins=[ages.min(), 16, 18, 20, 22, 24, ages.max()])
plt.show()
```

```python
plt.hist(ages, cumulative=True)
plt.show()
```

```python
# Pie chart
programming_languages = ['Python', 'Java', 'C++', 'Ruby', 'C#']
votes = [33, 17, 45, 10, 5]
explode = [0, 0, 0, 0, 0.3]
plt.pie(votes, labels=programming_languages,
        autopct='%1.2f%%', explode=explode)
plt.show()
```

```python
# multiple plots
stock_a = np.random.normal(0, 1, 10)
stock_b = np.random.normal(0, 1, 10)
stock_c = np.random.normal(0, 1, 10)
plt.plot(stock_a, c='r', label='stock_a')
plt.plot(stock_b, c='g', label='stock_b')
plt.plot(stock_c, c='b', label='stock_c')
plt.legend(loc='upper right')
plt.show()
```

```python
# subplots
x = np.arange(1, 50)
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, np.log(x))
axs[1, 0].set_title('log(x)')
axs[1, 1].plot(x, np.exp(x))
axs[1, 1].set_title('exp(x)')
fig.suptitle('four plots')
plt.show()
```
